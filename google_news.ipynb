```python
from pygooglenews import GoogleNews
import requests
from bs4 import BeautifulSoup

# Function to extract article content from a given URL
def extract_article_content(url):
    try:
        headers = {
            "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/107.0.0.0 Safari/537.36"
        }
        response = requests.get(url, headers=headers)
        response.raise_for_status()  # Raise an error for bad responses
        print(f"Successfully fetched: {url} (Status Code: {response.status_code})")

        soup = BeautifulSoup(response.content, 'html.parser')

        # Example: Adjust based on the website structure (this might need tweaking)
        paragraphs = soup.find_all('p')
        article_text = ' '.join([para.get_text() for para in paragraphs])

        if not article_text:
            print("No text found in the article.")
        return article_text.strip()

    except requests.exceptions.RequestException as e:
        print(f"Request error: {e}")
        return None
    except Exception as e:
        print(f"Failed to extract content from {url}: {e}")
        return None

# Search for articles
gn = GoogleNews()
search_results = gn.search('intitle:Devendra Fadnavis', when='10h')

print(search_results['feed'].title)

# Loop through the articles and extract content
for article in search_results['entries']:
    title = article.title
    link = article.link
    print(f"Title: {title}")
    print(f"Link: {link}")

    # Extract and print the article content
    content = extract_article_content(link)
    if content:
        print(f"Content:\n{content}\n")
    else:
        print("Content extraction failed.\n")
```
